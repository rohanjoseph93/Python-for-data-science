## Linear Bezier Curve

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib notebook
%matplotlib inline


# x axis values
x = [6,8]
# y axis values
y = [2,4]

# Establish the t values
t_values = np.arange(0,1.05,0.05)

# Bezier Curve for various values of t
x_poly = list()
y_poly = list()
for t in t_values:
    #print(t)
    x_poly.append((t*x[0] + (1-t)*x[1]))
    y_poly.append(t*y[0] + (1-t)*y[1])

fig, ax = plt.subplots()
# plotting the points 
plt.scatter(x, y,label = 'Linear')
plt.plot(x_poly, y_poly, label = 'Bezier Curve', color = 'green')

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

ax.text(6,2.1,"$P_0$")
ax.text(8,3.8,"$P_1$")
 
 
# function to show the plot
plt.show()
```
