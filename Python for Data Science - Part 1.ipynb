# 7. List comprehension

```python
#return even numbers from 1 to 100

even=[]
for i in range(100):
    if i%2 ==0:
        even.append(i)
    else:
        None
        
```

```python
print(even)
```

```python
#Let's do the same using list comprehension

even = [i for i in range(100) if i%2==0]
print(even)
```
